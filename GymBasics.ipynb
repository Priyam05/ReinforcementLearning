```python
import gym
```

```python
env=gym.make('CartPole-v0')   #Similarly try MountainCar-v0, MsPacman-v0, Hopper-v1
env.reset()

for _ in range(1000):
    env.render()
    #env.step(env.action_space.sample())
    observation, reward, done, info = env.step(env.action_space.sample())
    if done:
        print('Episode: {} finished'.format(_))
        env.reset()
env.close()
```

```python
env=gym.make('CartPole-v0')

for i_episode in range(20):
    env.reset()
    
    for t in range(1000):
        env.render()
        action=env.action_space.sample()
        observations, reward, done, info=env.step(action) #Observations is the state after taking the action
        print(info)
        if done:
            print('Episode:{0} finished at time step:{1}'.format(i_episode, t))
            break

env.close()
```
